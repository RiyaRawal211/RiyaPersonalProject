```python
# RENAMING COLUMN for standardization

df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
#Changing data types of columns
```

```python
df.dtypes
```

```python
#  purchase_amount (removed $ and commas, converted to float)

df['purchase_amount'] = df['purchase_amount'].replace(r'[\$,]', '', regex=True).astype(float)

```

```python
# Converted to datetime
df['time_of_purchase'] = pd.to_datetime(df['time_of_purchase'], errors='coerce')
```

```python
# Converted these columns to category
category_cols = [
    'gender', 'income_level', 'marital_status', 'education_level',
    'occupation', 'location', 'purchase_category', 'purchase_channel',
    'social_media_influence', 'discount_sensitivity', 'engagement_with_ads',
    'device_used_for_shopping', 'payment_method', 'purchase_intent',
    'shipping_preference'
]
for col in category_cols:
    df[col] = df[col].astype('category')
```

```python
# converted product_rating to float

df['product_rating'] = df['product_rating'].astype(float)
```

```python
#Standardized column names

df.columns = [col.title() for col in df.columns]

```

```python
#Standardized Categorical Values

df['Gender'] = df['Gender'].str.lower().str.strip()
df['Income_Level'] = df['Income_Level'].str.lower().str.strip()
df['Marital_Status'] = df['Marital_Status'].str.lower().str.strip()
df['Education_Level'] = df['Education_Level'].str.lower().str.strip()
df['Occupation'] = df['Occupation'].str.lower().str.strip()
df['Location'] = df['Location'].str.lower().str.strip()
df['Purchase_Category'] = df['Purchase_Category'].str.lower().str.strip()
df['Purchase_Channel'] = df['Purchase_Channel'].str.lower().str.strip()
df['Social_Media_Influence'] = df['Social_Media_Influence'].str.lower().str.strip()
df['Discount_Sensitivity'] = df['Discount_Sensitivity'].str.lower().str.strip()
df['Engagement_With_Ads'] = df['Engagement_With_Ads'].str.lower().str.strip()
df['Device_Used_For_Shopping'] = df['Device_Used_For_Shopping'].str.lower().str.strip()
df['Payment_Method'] = df['Payment_Method'].str.lower().str.strip()
df['Purchase_Intent'] = df['Purchase_Intent'].str.lower().str.strip()
df['Shipping_Preference'] = df['Shipping_Preference'].str.lower().str.strip()
```

```python
df.head()
```

```python
# Remove Duplicates

df.drop_duplicates(inplace=True)
```

```python
# fixing encoding issues
```

```python
import unicodedata

def normalize_text(text):
    if isinstance(text, str):
        # Normalize unicode, remove accents/diacritics
        text = unicodedata.normalize('NFKD', text)
        text = ''.join(c for c in text if not unicodedata.combining(c))
        # Optional: lowercase and strip
        return text.lower().strip()
    return text
```

```python
# applied to all

df['Location'] = df['Location'].apply(normalize_text)
```

```python
#HANDLING MISSING VALUES
```

```python
df.isnull().sum()         # seeing how many missing
```

```python
# since 25% values are missing : filling with 'unknown'

df['Social_Media_Influence'] = df['Social_Media_Influence'].fillna('unknown')
df['Engagement_With_Ads'] = df['Engagement_With_Ads'].fillna('unknown')
```

```python
df.isnull().sum()  
```

```python
# adding new columns

df['Purchase_Day'] = pd.to_datetime(df['Time_Of_Purchase'], errors='coerce').dt.day_name()
df['Is_Weekend'] = df['Purchase_Day'].isin(['Saturday', 'Sunday'])
```
